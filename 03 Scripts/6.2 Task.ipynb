# Categorical plots

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the price ranges. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

sns.histplot(df['price'], bins = 20, kde = True)
```

```python
df.loc[df['price'] < 400000, 'Price category'] = 'Low price'
```

```python
df.loc[(df['price'] >= 400000) & (df['price'] < 600000), 'Price category'] = 'Middle price'
```

```python
df.loc[df['price'] >= 600000, 'Price category'] = 'High price'
```

```python
df['Price category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="sqft_living", y="price", hue="Price category", data=df)
```

```python
# fig = g.get_figure()
g.savefig("categorical_plot.png") 
```

## The categorical plot again proves the positive correlation between sqft_living and price
